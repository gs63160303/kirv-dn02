```python
def gcd(a, b):
    while True:
        r = a % b
        if not r:
            return b
        a = b
        b = r
        pass
    pass
```

```python
import time
def avg_time(func, args_func, repetitions=1000):
    elapsed_times = []
    for _ in range(repetitions):
        args = args_func()
        start = time.time()
        func(*args)
        end = time.time()
        elapsed_times.append(end - start)
        pass
    return sum(elapsed_times) / len(elapsed_times)
```

```python
import random
def rand_tuple(n):
    return (lambda: (random.randrange(1 << (n - 1), 1 << n), random.randrange(1 << (n - 1), 1 << n)))
```

```python
n_from = 10
n_to = 10**3
step = 10
    
times = {n: avg_time(gcd, rand_tuple(n)) for n in range(n_from, n_to, step)}
# times
```

```python
import matplotlib.pylab as plt
x, y = zip(*times.items())
plt.plot(x, y)
plt.ylabel("Average time [seconds]")
plt.xlabel("Size of number [bits]")
plt.show()
```
